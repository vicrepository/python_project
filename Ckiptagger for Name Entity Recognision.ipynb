Function to print WS-POS-NER pipline results

```python
def print_word_pos_sentence(word_sentence, pos_sentence):
    assert len(word_sentence) == len(pos_sentence)
    for word, pos in zip(word_sentence, pos_sentence):
        print(f"{word}({pos})", end="\u3000")
    print()
    return
    
def print_ws_pos_ner_result(sentence_list, word_sentence_list, pos_sentence_list, entity_sentence_list):
    for i, sentence in enumerate(sentence_list):
        print()
        print(f"'{sentence}'")
        print_word_pos_sentence(word_sentence_list[i],  pos_sentence_list[i])
        for entity in sorted(entity_sentence_list[i]):
            print(entity)
```

```python
print_ws_pos_ner_result(sentence_list, word_sentence_list, pos_sentence_list, entity_sentence_list)
```

Use the entity list to extract countries mentioned in the title and the overall count

```python
from collections import Counter

# Use entity_sentence_list'GPE', 'PERSON', 'NORP','FAC', 'ORG', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE', 'DATE', 'TIME', 'PERCENT', 'MONEY', 'QUANTITY', 'ORDINAL', 'CARDINAL'
# Flatten the list of entities
flat_entities = [e[3] for sublist in entity_sentence_list for e in sublist if e[2] == 'GPE']

# Use the Counter function to count the occurrences of each country
country_counts = Counter(flat_entities)

# Get the countries sorted by frequency
sorted_countries = sorted(country_counts.items(), key=lambda x: x[1], reverse=True)

# Print the countries and their counts
print("Countries and their counts:")
for country, count in sorted_countries:
    print(country, count)
```
